### Data Cleaning

```python
df.isnull().sum()
```

```python
df.drop(columns = ['state','comments'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df['self_employed'] = df['self_employed'].fillna('No')
df['work_interfere'] = df['work_interfere'].fillna("Don't Know")
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df['Gender'].value_counts()
```

```python
male = ['M', 'Male', 'male', 'm', 'Male-ish', 'maile', 'Cis Male', 'Mal', 'Male (CIS)', 'Make', 'Man', 'msle', 'Mail', 'cis male']

female = ['Female', 'female', 'Cis Female', 'F', 'Woman', 'f', 'Femake', 'woman', 'cis-female/femme', 'Female (cis)', 'femail']

df['Gender'] = df['Gender'].replace(male, 'Male')
df['Gender'] = df['Gender'].replace(female, 'Female')
df['Gender'] = df['Gender'].apply(lambda x: 'Other' if x not in ['Male', 'Female'] else x)
```

```python
df['Gender'].value_counts()
```

```python
df.duplicated().sum()
```

```python
df['Age'].value_counts()
```

```python
df[df['Age'] < 18].shape[0]
```

```python
df[df['Age'] > 100].shape[0]
```

```python
df = df[(df['Age'] > 18) & (df['Age'] < 100)]
```

```python
df['age_group'] = pd.cut(df['Age'], bins=[18,24,34,44,54,65], labels=['18-24', '25-34', '35-44', '45-54', '55-65'])
df['age_group'].value_counts()
```

#### Observation

First, I checked for missing values across all columns and found many, especially in State and Comments, so I removed those rows completely. Next, I handled missing values in the Self-Employed and Work Interfere columns. I then cleaned the messy Gender column, standardizing it to 'Male' (986 people), 'Female' (245 people), and 'Other' (28 people). After dropping duplicates, I removed invalid ages like negatives or values over 100 using boolean indexing and created age groups. The 25-34 group is the biggest at over 707 people, while 55–65 has the fewest at just 16.

### Exploratory Data Analysis

#### Treatment Distribution

```python
sns.countplot(data=df, x='treatment', hue='treatment', palette='muted')
plt.title('Treatment distribution')
plt.xlabel('treatment')
plt.ylabel('count')
plt.show()
```

```python
df['treatment'].value_counts(normalize=True) * 100
```

#### Observation

I used a count plot as it's used for counts of observations in categorical data. The number of people who have taken treatment vs. those who haven't is almost similar, with nearly 51 percent having sought treatment for a mental health condition and 49 percent not.

#### Age vs Treatment

```python
sns.boxplot(data=df, x='treatment', y='Age', palette='muted')
plt.title('Age distribution by treatment')
plt.show()
```

```python
df.groupby('age_group')['treatment'].value_counts(normalize=True) * 100
```

#### Observation

In this analysis, I used a box plot to see the distributions of numeric data across groups. Ages 18-34 are highest in number seeking treatment vs other groups, but percentages within each (taking vs not) are nearly same (around 50%), except 55-65 group where almost 68% seek it.

#### Gender vs Treatment

```python
sns.countplot(data=df, x='Gender', hue='treatment', palette='muted')
plt.title('Gender vs Treatment')
plt.xlabel('gender')
plt.ylabel('count')
plt.show()
```

```python
df.groupby('Gender')['treatment'].value_counts(normalize=True) * 100
```

#### Observation

In females, more than 68% seek treatment, whereas in males it's slightly above half (54%), and 79% in the other group.

#### Family history & work interference vs Treatment

```python
fig, axes = plt.subplots(1,2, figsize=(14,5))
pd.crosstab(df['family_history'], df['treatment']).plot(kind='bar', stacked=True, ax=axes[0], colormap='Set2')
axes[0].set_title('family history vs treatment')
axes[0].set_xlabel('family_history')

pd.crosstab(df['work_interfere'], df['treatment']).plot(kind='bar', stacked=True, ax=axes[1], colormap='Set2')
axes[1].set_title('work interference vs treatment')
axes[1].set_xlabel('work interference')

plt.tight_layout()
plt.xticks(rotation=30)
plt.show()
```

#### Observation

People with family history of mental health issues seek treatment at much higher rates. Work interference shows more even distribution, with most reporting "sometimes."

#### Company Size vs Treatment

```python
df.groupby('no_employees')['treatment'].value_counts(normalize=True).unstack().plot(kind='barh', colormap='Set2',figsize=(10,4))
plt.title('company size vs treatment')
plt.xlabel('proportion')
plt.show()
```

#### Observation

The proportion seeking treatment stays relatively consistent above 0.4 regardless of company size.

#### Top 10 countries

```python
top10= df['Country'].value_counts().head(10)
top10.plot(kind='bar', colormap='Set2', figsize=(10,5))   
plt.title('Top 10 countries')
plt.xlabel('country')
plt.ylabel('count')
plt.xticks(rotation=30)
plt.show()
```

#### Observation

USA leads the list with a huge gap over the next countries, followed by UK and Canada.

#### Correlation Heatmap

```python
df_encoded=df.copy()
df_encoded=df_encoded.drop(columns=['age_group'])
for col in df_encoded.select_dtypes('object').columns: df_encoded[col]= df_encoded[col].astype('category').cat.codes

plt.figure(figsize=(14,10))
sns.heatmap(df_encoded.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation matrix')
plt.show()
```

#### Observation

No single factor dominates, as relationships are distributed across all variables showing only weak-to-moderate correlations.
